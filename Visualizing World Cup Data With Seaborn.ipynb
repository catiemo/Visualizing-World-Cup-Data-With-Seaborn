### Total number of goals in each match

```python
goals_per_year = df.groupby('Year')['Total Goals'].sum().reset_index()
goals_per_year
```

### Bar chart of goals per tournament

```python
sns.set_style('whitegrid')
sns.set_context('poster', font_scale = 0.5)
f, ax = plt.subplots(figsize = (12, 7))
ax = sns.barplot(data = goals_per_year, x = 'Year', y = 'Total Goals')
ax.set_title('Goals Per Tournament')
plt.xticks(rotation = 60)
plt.show()
plt.clf()
```

### Boxplot of Goals Per Tournament

```python
sns.set_style('darkgrid')
sns.set_context('notebook', font_scale = 1.25)
f, ax2 = plt.subplots(figsize = (12, 7))
ax2 = sns.boxplot(data = df_goals, x = 'year', y = 'goals', palette = 'Spectral')
ax2.set_title('Distribution of Goals Per Tournament')
plt.xticks(rotation = 60)
plt.show()
```
